```python
import pandas as pd
```

```python
# Load data
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/udemy_courses.csv')
```

```python
df.head()
```

```python
# Preprocessing published_timestamp data (2017-01-18T20:58:58Z) to get Y-m-d format (2017-01-18)
df['published_timestamp'] = df['published_timestamp'].str.split('T', expand=True)[0]
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
# Get URL list from dataset
urls = df['url']

# Create an array to store URLs that can not be accessed
urls_to_remove = []

for url in urls:
  try:
      response = requests.get(url)
  except requests.exceptions.ChunkedEncodingError as e:
      urls_to_remove.append(url)
  except requests.exceptions.RequestException as e:
      urls_to_remove.append(url)
  else:
    if response.status_code == 200:
      soup = BeautifulSoup(response.text, 'html.parser')
      topic_menu = soup.find('div', class_='topic-menu')

      if topic_menu:
        category = topic_menu.findAll('a')[0]
        sub_category = topic_menu.findAll('a')[1]
        topic = topic_menu.findAll('a')[2]

        df.loc[df['url'] == url, 'category'] = category.text.strip()
        df.loc[df['url'] == url, 'sub_category'] = sub_category.text.strip()
        df.loc[df['url'] == url, 'topic'] = topic.text.strip()

df.head()
```

```python
# Write output to new file
df.to_csv('udemy_topic_data.csv', index=False)
```
